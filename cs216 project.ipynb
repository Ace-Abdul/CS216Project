```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df2015 = pd.read_csv('2015.csv')
df2016 = pd.read_csv('2016.csv')
df2017 = pd.read_csv('2017.csv')
df2018 = pd.read_csv('2018.csv')
df2019 = pd.read_csv('2019.csv')
df2020 = pd.read_csv('2020.csv')
df2021 = pd.read_csv('2021.csv')
df2022 = pd.read_csv('2022.csv')
df2020.head()
```

```python
dfgini = pd.read_csv('Gini Coefficient.csv')
dfgini.head(10)
```

```python
merged_df = pd.merge(df2020, dfgini, left_on='Country name', right_on='country', how='left')
merged_df.head()
```

```python
sns.set(rc={'figure.figsize':(15,8),'axes.facecolor':'#E7D6CC', 'figure.facecolor':'white'})
color=sns.color_palette("Paired")
color
```

```python
sns.boxplot(x=merged_df['Ladder score'],y=merged_df['Regional indicator'], palette=color)
```

```python
columns = ['Logged GDP per capita', 'Social support', 'Healthy life expectancy', 'Freedom to make life choices', 'Generosity', 'Perceptions of corruption']


scols = int(len(columns)/2)
srows = 2
fig, axes = plt.subplots(scols, srows, figsize=(10,6))

for i, col in enumerate(columns):
    ax_col = int(i%scols)
    ax_row = int(i/scols)
    
    sns.distplot(merged_df[col], hist=True, ax=axes[ax_col, ax_row])
    axes[ax_col, ax_row].set_title('Frequency distribution '+ col, fontsize=12)
    axes[ax_col, ax_row].set_xlabel(col, fontsize=8)
    axes[ax_col, ax_row].set_ylabel('Contagem', fontsize=8)
fig.tight_layout()
plt.show()
```

```python
columns = ['Logged GDP per capita', 'Social support', 'Healthy life expectancy', 'Freedom to make life choices', 'Generosity',\
           'Perceptions of corruption']
sns.pairplot(merged_df[columns])
```

```python
numeric_cols = merged_df.select_dtypes(include=np.number)
numeric_cols.nunique(axis=0)
numeric_cols.describe().apply(lambda s: s.apply(lambda x: format(x, 'f')))
def describe(df, stats):
    d = df.describe()
    return d.append(df.reindex(d.columns, axis=1).agg(stats))
describe(numeric_cols, ['skew', 'mad', 'kurt'])
```

```python
missing_values = numeric_cols.isnull().sum()
print(missing_values)
```

```python
corr = numeric_cols.corr()# plot the heatmap
sns.heatmap(corr, xticklabels=corr.columns, yticklabels=corr.columns, annot=True, cmap=sns.diverging_palette(220, 20, as_cmap=True))
```
